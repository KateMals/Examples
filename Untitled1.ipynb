```python
# Показать цену бриллиантов, каратность которых больше 4 карат
dfD.loc[dfD['carat'] > 4, ['price', 'carat']]
```

```python
# Максимальное значение цены
print(dfD['price'].max())
# Минимальное значение цены
print(dfD['price'].min())
# Среднее значение цены
print(dfD['price'].mean())
# Сумма цен
print(dfD['price'].sum())
```

```python
# Максимальное значение карата
print(dfD['carat'].max())
# Минимальное значение карата
print(dfD['carat'].min())
# Среднее значение карата
print(dfD['carat'].mean())
# Сумма карат
print(dfD['carat'].sum())
```
